```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time):
  a = 0.1
  b = -1
  c = 30

  temperature = a * (time**2) + b * time + c
  return temperature

def main():
   time_values = np.linspace(0,10,50)

   temperature_harcoded = quadratic_model(time_values)

   plt.plot(time_values,  temperature_harcoded , label='Hard-coded Coefficients')
   plt.xlabel('Time')
   plt.ylabel('Temperature')
   plt.legend()
   plt.title('Weather Modelling with Quadratic Equation (Hard-Coded Coefficients)')
   plt.show()

if __name__ == "__main__":
 main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time):
  a = float(input("Enter a value : "))
  b = float(input("Enter b value : "))
  c = float(input("Enter c value : "))

  temperature = a * (time**2) + b * time + c
  return temperature

def main():
   time_values = np.linspace(0,10,50)

   temp = quadratic_model(time_values)

   plt.plot(time_values,  temp , label=f'User Input Coefficients')
   plt.xlabel('Time')
   plt.ylabel('Temperature')
   plt.legend()
   plt.title('Weather Modelling with Quadratic Equation (User- Input Coefficients)')
   plt.show()

if __name__ == "__main__":
 main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, coeffs):
    a = float(coeffs[0])
    b = float(coeffs[1])
    c = float(coeffs[2])
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    with open("exp3.txt", 'r') as f:
        lines = f.readlines()

    for idx, line in enumerate(lines):
        coeffs = line.strip().split(",")
        temp = quadratic_model(time_values, coeffs)
        plt.plot(time_values, temp, label=f'Set {idx + 1} {coeffs}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Input from file)')
    plt.show()

if __name__ == "__main__":
 main()
```
